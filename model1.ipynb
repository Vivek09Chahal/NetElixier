```python
import pandas as pd
import numpy as np

import tensorflow as tf

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Train and Evaluation Data
google_data = pd.read_csv('Train Data/googleads-performance.csv')
meta_data = pd.read_csv('Train Data/metaads-performance.csv')
microsoft_data = pd.read_csv('Train Data/microsoftads-performance.csv')
```

```python
data = pd.concat([google_data, microsoft_data], axis=0)
```

```python
data['Clicks'] = pd.to_numeric(data['Clicks'], errors='coerce')
```

```python
data.info()
```

```python
data = data.drop(['Date', 'Campaign type'], axis=1)
```

```python
data.count()
```

```python
# Setting X and Y Values

X = data.drop(['Revenue', 'Conversions'], axis=1)
y = data[['Revenue', 'Conversions']]
```

```python
data.info()
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mse,
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
                metrics=['mse'])

model.fit(X, y, epochs=100)
```
